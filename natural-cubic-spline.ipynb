The **natural cubic spline** is a piecewise polynomial fitting procedure. 
- It uses polynomials with degree $3$, i.e. 'cubic'. It is claimed that cubic splines are the lowest-order spline for which the knot-discontinuity is not visible to the human eye. There is seldom any good reason to go beyond cubic-splines. 
- Impose the constraint that it is twice differentiable on the knot points
- It is linear for extrapolation. This frees up four degrees of freedom (two constraints each in both boundary regions), which can be spent more profitably by sprinkling more knots in the interior region. This tradeoff is illustrated in terms of variance in the figure below. There will be a price paid in bias near the boundaries, but assuming the function is linear near the boundaries (where we have less information anyway) is often considered reasonable.

A natural cubic spline with K knots is represented by $K$ basis functions.
\begin{align}
N_1(X)=1,\;\;N_2(X)=X, N_{k+2}(X)=d_k(X)-d_{K-1}(X),
\end{align}
where
\begin{align}
d_k(X)=\frac{(X-\xi_k)_+^3-(X-\xi_K)_+^3}{\xi_K-\xi_k}.
\end{align}


The polynomials fit beyond the boundary knots behave even more wildly than the corresponding global polynomials in that region; see the graph below.
